````python
# import libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
````

````python
# read the CSV file, 'iris.csv'
iris = pd.read_csv('./iris.data', header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])
iris.head()
````

````python
# show total number of rows and columns
print(iris.shape)
````

````python
# show all the column names
print(iris.columns)
````

````python
# count how many different 'type of answers' the table has
iris['species'].value_counts()
````

This is an example of a 'data-set' that is 'balance' (you can split them equally)

#### Now we are going to 'plot' the data into a 'scatter plot' (graph)

````python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
````

as you can see, the each dot is one 'sample' and the dots are all over the place; so, you can't\
really 'drawn' any conclusion about the 'data-set'

so we are going to try something different now

````python
sns.set_style("whitegrid") # this line must comes FIRST
sns.FacetGrid(iris, hue='species', size=4).map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()


plt.show()
````

Now we are going to do a 3 dimension 'scatter plot'
https://plot.ly/pandas/3d-scatter-plots/

need mouse clicking to move the 'scatter plot' around

````python
plt.close()

sns.set_style('whitegrid')
sns.pairplot(iris, hue='species', height=3)
plt.show()
````
